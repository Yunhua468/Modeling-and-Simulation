```python
import random
from collections import deque
```

```python
class Customer(object):
  def __init__(self, arrival_time, service_time):
    self.arrival_time = arrival_time
    self.service_time = service_time
    self.service_begin = -1
    self.waiting_time = -1

  def __repr__(self):
    return '[arrival_time: {}, service_time: {}, service_begin: {}, waiting_time: {}]'.format(self.arrival_time, 
                                                                                            self.service_time, 
                                                                                            self.service_begin, 
                                                                                            self.waiting_time)
```

```python
customers = deque()
time = 0
for i in range(1000):
  time = time + random.randint(8, 12)
  customer = Customer(arrival_time=time, service_time=random.randint(10, 16))
  customers.appendleft(customer)
  print(customer)
```

```python
queues = [deque()]
served = []
x = 10
for time in range(0, 10010):

  if len(customers) == 0:
    break

  if (len(queues[-1]) > x) and (len(queues) <= 6):
    print('open new register')
    queues.append(deque())


  # Check arrivals and add to the correct line
  if customers[-1].arrival_time == time:
    for qi, queue in enumerate(queues):
      if (len(queue) < x) or (qi == (len(queues) - 1)):
        customer = customers.pop()
        queue.appendleft(customer)
        print('{}:: added customer: {} to queue {}'.format(time, customer, qi))
        break
  
  for queue in queues:
    if len(queue) == 0:
      continue

    if queue[-1].service_begin == -1:
      # start service of the customer
      print('')
      queue[-1].service_begin = time
    
    if (queue[-1].service_begin + queue[-1].service_time) == time:
      customer = queue.pop()
      customer.waiting_time = time - customer.arrival_time
      served.append(customer)
      print('{}:: serviced: {}'.format(time, customer))
    
```

```python
s = 0
for customer in served:
  s += customer.waiting_time
print(s / len(served))
```
